```python
%pylab inline


pylab.rcParams['figure.figsize'] = (14.0, 4.0)

import numpy as np
import matplotlib.pyplot as plt

# Get the IMU signals from a file
def getSigs(fileName):
    fin = open(fileName)

    lines = fin.readlines()
    fin.close()
    
    res = []
    for str in lines:
        tmp = str.split()
        tmp = map(eval, tmp)
        res.append(tmp)

    res_array = np.array(res)
    res_array = res_array[10:, :]
    correctBias(res_array)
    
    return res_array



def correctBias(array):
    #still_array = getSigs('../data/still_normal.txt')
    
    
    # TODO, the parameters may need be adjusted when in different posture
    # The bias of gyr_x, gyr_y, gyr_z, statistics from still_normal.txt
    gyr_bias = np.array([-0.01568079, 0.02071555, 0.0186753 ])
    array[:, 3:6] = array[:, 3:6] - gyr_bias

    
    
def getSpectrum(array):
    N = len(array)
    frame_len = 30
    
    
    frame_num = N/frame_len
    #spectrum = np.zeros((frame_len/2, frame_num+1))
    #print spectrum.shape
    spectrum = []
    
    fft_len = 32
    
    for i in range(0, N, frame_len):
        array_seg = array[i : i+frame_len]
        spectrum.append(np.log(np.abs(np.fft.fft(array_seg, fft_len)))[:fft_len/2])
        
    return np.array(spectrum)


def plotSpectrum(array, spectrum, type):
    fig, ax = plt.subplots(2)
    ax[0].plot(array)
    ax[0].set_xlabel('time')
    ax[0].set_title(type)
    
    
    
    ax[1].pcolor(spectrum)
    ax[1].set_xlabel('frame id')
    ax[1].set_ylabel('frequency')
    
    plt.show()
    

# Calculate the energy
def energyXYZ(array):
    energy_acc = np.sqrt(np.sum(array[:,0:3]**2, 1))
    energy_gyr = np.sqrt(np.sum(array[:,3:6]**2, 1))
    energy_mag = np.sqrt(np.sum(array[:,6:]**2, 1))
    tmp = np.array([energy_acc, energy_gyr, energy_mag])
    return np.transpose(tmp)


def calEntrophy(gyr_energy):
    gyr_energy_norm = gyr_energy/sum(gyr_energy)
    entropy = -sum(gyr_energy_norm * np.log(gyr_energy_norm))
    return entropy


# 根据事件的时间长度和整个事件的能量熵来同时来检测是否有震动
def detectVibration(gyr_energy):
    # Energy uppper threshold
    T_upper = 0.03
    # Energy lower threshold
    T_lower = 0.01
    
    energy_max = max(gyr_energy)
    
    N = len(gyr_energy)
    start_ind = -1
    end_ind = N-1
    
    max_period = 0
    
    for i in range(N):
        if start_ind == -1 and gyr_energy[i] >= max(T_upper, energy_max/2):
            start_ind = i
        elif start_ind != -1 and (gyr_energy[i] < T_lower or i == N-1):
            end_ind = i
            tmp = end_ind - start_ind
            start_ind = -1
            if tmp > max_period:
                max_period = tmp
            
    entrophy = calEntrophy(gyr_energy)
    
    T_period = 10
    T_entrophy = 2.8
    # 熵很小，表示能量很集中
    if max_period < T_period and entrophy < T_entrophy:
        return True
    else:
        return False
            
            
    

if __name__ == '__main__':
    dir = '../data/'
    
    fileNames = ['still_normal.txt', 'still_frontDown.txt' ,'motion_xyz_Move_uniform1.txt',\
                 'motion_xyz_Move_uniform2.txt',\
                 'motion_xyz_Move_uniform3.txt', 'motion_xyz_Move_uniform4.txt',\
                 'motion_xRotate_uniform_anticlock.txt', 'motion_xRotate_uniform_clock.txt',\
                 'motion_vibration1.txt', 'motion_vibration2.txt']
    for tmp in fileNames:
        fileName = dir + tmp
    
        res_array = getSigs(fileName)
        
        energy_array = energyXYZ(res_array)
    
        types = ['-acc_x', '-acc_y', '-acc_z', '-gyr_x', '-gyr_y', '-gyr_z', '-mag_x', '-mag_y', '-mag_z']
        types_energy = ['-acc_energy', '-gyr_energy', '-mag_energy']
        
        for i in range(2):
            spectrum = getSpectrum(energy_array[:,i])
            if i == 1:
                print spectrum
            spectrum = np.transpose(spectrum)
            plotSpectrum(energy_array[:,i], spectrum, fileName + types_energy[i])
            
        
        interval = 30
        gyr_energy = energy_array[:,1]
        N = len(gyr_energy)
        
        vibrationRes = []
        for i in range(0, N, interval):
            gyr_energy_seg = gyr_energy[i: i+interval]
            existVibration = detectVibration(gyr_energy_seg)
            if existVibration == True:
                vibrationRes.append(1)
            else:
                vibrationRes.append(0)
                
        plt.plot(vibrationRes)
        plt.show()
        
        
        
        
        
    
```
